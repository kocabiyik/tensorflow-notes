# Data
Data downloaded from https://console.cloud.google.com/storage/quickdraw_dataset/full/binary

```python
data_dir = './data/quickdraw/'
files_list = [file for file in os.listdir(data_dir) if 'npy' in file]
```

```python
animals_dict = {
        'snake' : 0, 
    'butterfly' : 1, 
          'dog' : 2, 
         'bird' : 3, 
      'giraffe' : 4, 
          'bee' : 5, 
       'rabbit' : 6, 
     'elephant' : 7, 
          'ant' : 8, 
          'cat' : 9
}
```

```python
SAMPLE_SIZE_FOR_EACH = 100

animals = []
for file in files_list:
    draws = np.load(data_dir+file)
    a = file.replace('.npy', '').rsplit('_',1)[1]
    choices = np.random.randint(0,draws.shape[0], SAMPLE_SIZE_FOR_EACH)
    s = draws[choices]
    try:
        samples = np.append(samples, s,axis=0)
    except:
        samples = s
    animals.append(a)
```

```python
y = [animals_dict[i] for i in animals for x in range(SAMPLE_SIZE_FOR_EACH)]
```

```python
import matplotlib.pyplot as plt

n = 400
plt.figure(figsize=(15, 15))

for i in range(n):
    ax = plt.subplot(20, 20, i+1)
    choice = np.random.randint(0,samples.shape[0], 1)
    plt.imshow(samples[choice].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
